```python
#Linear Regression (for with 2020 data)

#Reshape into an array
x = airport.iloc[:, 0].values.reshape(-1, 1)
y = airport.iloc[:, 1].values.reshape(-1, 1)

print(x)
print(y)

#Predict passenger boardings for these years (independent variables) and reshape into an array
predict_x = [2021, 2022, 2023, 2024]
predict_x = np.array(predict_x).reshape(-1,1)

#Using linear regression
regres = LinearRegression()
regres.fit(x, y)

#y is our dependent variable and yhat is the predicted number of passenger boardings for these years
yhat = regres.predict(predict_x)

#Result
print('Predicted passenger boardings -> \n',yhat)

#Other...
slope = regres.coef_
intercept = regres.intercept_
```

```python
#For with 2020 data

%matplotlib inline 

#Produce a scatter plot 
plt.scatter(x,y, color = 'red')

plt.title('Predicted number of passengers', fontweight = 'bold')
plt.xlabel('Year')
plt.ylabel('Passengers')

#newgraph is the extrapolation based on the predicted values. It is the fitted values for passenger boardings at LAX
newgraph = [slope*i + intercept for i in np.append(x, predict_x)]
newgraph = np.array(newgraph).reshape(-1,1)
plt.plot(np.append(x,predict_x), newgraph, color = 'limegreen')
plt.show()
```

```python
#Delivery rate for short-range aircraft 
#
#Implemented through a linked list

class node:
    def __init__(self, data = None):
        self.data = data 
        self.next = None 

class sra:
    def __init__(self):
        self.head = None 

    def showdeliveryrate(self):
        if self.head == None:
            return None

        rates = [str(round(((1172/3857) * 100), 2)) + '%', str(round(((481/3437) * 100), 2)) + '%', str(round(((435/    (2135 + 234)) * 100), 2)) + '%']

        traverse = self.head 
        while traverse != None:
            for i in range(3):
                print(traverse.data + rates[i])
                traverse = traverse.next

smallplanes = sra()

smallplanes.head = node('A320neo -> ')
smallplanes1 = node('A321neo -> ')
smallplanes2 = node('737 MAX -> ')

smallplanes.head.next = smallplanes1 
smallplanes1.next = smallplanes2 

smallplanes.showdeliveryrate()
```

```python
#Delivery rate for long-range aircraft 
#
#Implemented through a linked list

class node:
    def __init__(self, data = None):
        self.data = data 
        self.next = None 

class sra:
    def __init__(self):
        self.head = None 

    def showdeliveryrate(self):
        if self.head == None:
            return None

        rates = [str(round(((355/745) * 100), 2)) + '%', str(round(((53/168) * 100), 2)) + '%', str(round(((47/155) * 100), 2)) + '%', str(round(((0/312) * 100), 2)) + '%']

        traverse = self.head 
        while traverse != None:
            for i in range(4):
                print(traverse.data + rates[i])
                traverse = traverse.next

smallplanes = sra()

smallplanes.head = node('A350-900 -> ')
smallplanes1 = node('A350-1000 -> ')
smallplanes2 = node('747-8i -> ')
smallplanes3 = node('777-9X -> ')

smallplanes.head.next = smallplanes1 
smallplanes1.next = smallplanes2
smallplanes2.next = smallplanes3

smallplanes.showdeliveryrate()
```
